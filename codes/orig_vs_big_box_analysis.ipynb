```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dpi = 100
```

```python
orig_assess = pd.read_excel('cube_cell_size_assess_orig-sim-box.xlsx', index_col = 0)
orig_assess
```

```python
big_assess = pd.read_excel('cube_cell_size_assess_big-sim-box.xlsx', index_col = 0)
big_assess
```

```python
for a in ['x', 'y', 'z']:

    plt.figure(figsize = (10, 4), dpi = dpi, facecolor=(1, 1, 1))
    width = 0.25
    plt.bar(orig_assess.index - width/2, orig_assess[f'r - V{a}'], width, label = '(640 h$^{-1}$ Mpc)$^3$')
    plt.bar(big_assess.index  + width/2, big_assess[f'r - V{a}'],  width, label = '(1920 h$^{-1}$ Mpc)$^3$');
    plt.ylim(0.3,1)
    plt.ylabel("Pearson's r")
    plt.legend(loc = 'upper left')
    plt.xticks(ticks = orig_assess.index, labels = [f'C: {cell}, B: {box}' for cell, box in zip(orig_assess['Cell Size'], orig_assess['Act Cube Size'])], 
               rotation = 60, fontsize = 6, horizontalalignment = 'right');
    plt.xlabel('Cell and Box Sizes (h$^{-1}$ kpc)')
    plt.title("Original vs Big Simulation Box | Pearson's r for V$_{}$".format(a))

    net = big_assess[f'r - V{a}'] - orig_assess[f'r - V{a}']
    plt.text(0.8, 0.9, f'More r cases: {len(net[net>0])}', transform=plt.gca().transAxes)
    plt.text(0.8, 0.85, f'Less r cases: {len(net[net<0])}', transform=plt.gca().transAxes)
```

```python
for a in ['x', 'y', 'z']:

    plt.figure(figsize = (10, 4), dpi = dpi, facecolor=(1, 1, 1))
    width = 0.25
    plt.bar(orig_assess.index - width/2, orig_assess[f'SD - V{a}'], width, label = '(640 h$^{-1}$ Mpc)$^3$')
    plt.bar(big_assess.index  + width/2, big_assess[f'SD - V{a}'],  width, label = '(1920 h$^{-1}$ Mpc)$^3$');
    plt.ylim(0,1200)
    plt.ylabel("SD of error in velocity (km/s)")
    plt.legend(loc = 'upper left')
    plt.xticks(ticks = orig_assess.index, labels = [f'C: {cell}, B: {box}' for cell, box in zip(orig_assess['Cell Size'], orig_assess['Act Cube Size'])], 
               rotation = 60, fontsize = 6, horizontalalignment = 'right');
    plt.xlabel('Cell and Box Sizes (h$^{-1}$ kpc)')
    plt.title("Original vs Big Simulation Box | SD of error in V$_{}$".format(a))

    net = orig_assess[f'SD - V{a}'] - big_assess[f'SD - V{a}']
    plt.text(0.8, 0.90, f'Less SD cases: {len(net[net>0])}', transform=plt.gca().transAxes)
    plt.text(0.8, 0.85, f'More SD cases: {len(net[net<0])}', transform=plt.gca().transAxes)
```
