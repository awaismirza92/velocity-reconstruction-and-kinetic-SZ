**Importing libraries**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

**Setting notebook parameters**

```python
dpi = 100
```

**Reading & printing the assessment file for the original simulation box**

```python
orig_assess = pd.read_excel('cube_cell_size_assess_orig-sim-box.xlsx', index_col = 0)
orig_assess
```

**Reading & printing the assessment file for the big simulation box**

```python
big_assess = pd.read_excel('cube_cell_size_assess_big-sim-box.xlsx', index_col = 0)
big_assess
```

**Plotting the bar charts to compare Pearson'r for both the boxes**

```python
for a in ['x', 'y', 'z']:

    plt.figure(figsize = (10, 4), dpi = dpi, facecolor=(1, 1, 1))
    width = 0.25
    plt.bar(orig_assess.index - width/2, orig_assess[f'r - V{a}'], width, label = '640$^3$ h$^{-3}$Mpc$^3$')
    plt.bar(big_assess.index  + width/2, big_assess[f'r - V{a}'],  width, label = '1920$^3$ h$^{-3}$Mpc$^3$');
    plt.ylim(0.3,1)
    plt.ylabel("Pearson's r")
    plt.legend(loc = 'upper left')
    plt.xticks(ticks = orig_assess.index, labels = [f'C: {cell}, B: {box}' for cell, box in zip(orig_assess['Cell Size'], orig_assess['Act Cube Size'])], 
               rotation = 60, fontsize = 6, horizontalalignment = 'right');
    plt.xlabel('Cell and Box Sizes (h$^{-1}$ kpc)')
    plt.title("Original vs Big Simulation Box | Pearson's r for V$_{}$".format(a))

    net = big_assess[f'r - V{a}'] - orig_assess[f'r - V{a}']
    plt.text(0.8, 0.93, f'More r cases: {len(net[net>0])}', transform=plt.gca().transAxes)
    plt.text(0.8, 0.87, f'Less r cases: {len(net[net<0])}', transform=plt.gca().transAxes)
    plt.text(0.8, 0.81, f'Average change: {round(net.mean(),3)}', transform=plt.gca().transAxes)
    plt.tight_layout()
    plt.savefig(f'Plots/orig-vs-big-box/Pearson_r/V{a}.png')
```

**Plotting the bar charts to compare the standard deviation in the velocity error for both the boxes**

```python
for a in ['x', 'y', 'z']:

    plt.figure(figsize = (10, 4), dpi = dpi, facecolor=(1, 1, 1))
    width = 0.25
    plt.bar(orig_assess.index - width/2, orig_assess[f'SD - V{a}'], width, label = '640$^3$ h$^{-3}$Mpc$^3$')
    plt.bar(big_assess.index  + width/2, big_assess[f'SD - V{a}'],  width, label = '1920$^3$ h$^{-3}$Mpc$^3$');
    plt.ylim(0,1200)
    plt.ylabel("SD of error in velocity (km/s)")
    plt.legend(loc = 'upper left')
    plt.xticks(ticks = orig_assess.index, labels = [f'C: {cell}, B: {box}' for cell, box in zip(orig_assess['Cell Size'], orig_assess['Act Cube Size'])], 
               rotation = 60, fontsize = 6, horizontalalignment = 'right');
    plt.xlabel('Cell and Box Sizes (h$^{-1}$ kpc)')
    plt.title("Original vs Big Simulation Box | SD of error in V$_{}$".format(a))

    net = big_assess[f'SD - V{a}'] - orig_assess[f'SD - V{a}']
    plt.text(0.75, 0.93, f'Less SD cases: {len(net[net<0])}', transform=plt.gca().transAxes)
    plt.text(0.75, 0.87, f'More SD cases: {len(net[net>0])}', transform=plt.gca().transAxes)
    plt.text(0.75, 0.81, f'Average change: {round(net.mean(),1)} km/s', transform=plt.gca().transAxes)
    plt.tight_layout()
    plt.savefig(f'Plots/orig-vs-big-box/SD_of_error/V{a}.png')
```

**Plotting the bar charts to compare the absolute value of the mean of velocity error for both the boxes**

```python
for a in ['x', 'y', 'z']:

    plt.figure(figsize = (10, 4), dpi = dpi, facecolor=(1, 1, 1))
    width = 0.25
    plt.bar(orig_assess.index - width/2, abs(orig_assess[f'Mean - V{a}']), width, label = '640$^3$ h$^{-3}$Mpc$^3$')
    plt.bar(big_assess.index  + width/2, abs(big_assess[f'Mean - V{a}']),  width, label = '1920$^3$ h$^{-3}$Mpc$^3$');
    plt.ylim(0, 1000)
    plt.ylabel("Mean error in velocity (km/s)")
    plt.legend(loc = 'upper left')
    plt.xticks(ticks = orig_assess.index, labels = [f'C: {cell}, B: {box}' for cell, box in zip(orig_assess['Cell Size'], orig_assess['Act Cube Size'])], 
               rotation = 60, fontsize = 6, horizontalalignment = 'right');
    plt.xlabel('Cell and Box Sizes (h$^{-1}$ kpc)')
    plt.title("Original vs Big Simulation Box | Mean error for V$_{}$".format(a))

    net = abs(big_assess[f'Mean - V{a}']) - abs(orig_assess[f'Mean - V{a}'])
    plt.text(0.75, 0.93, f'Less abs v cases: {len(net[net<0])}', transform=plt.gca().transAxes)
    plt.text(0.75, 0.87, f'More abs v cases: {len(net[net>0])}', transform=plt.gca().transAxes)
    plt.text(0.75, 0.81, f'Average change: {round(net.mean(),1)} km/s', transform=plt.gca().transAxes)
    plt.tight_layout()
    plt.savefig(f'Plots/orig-vs-big-box/Abs_mean_of_error/V{a}.png')
```

**Plotting the bar charts to compare the signed value of the mean of velocity error for both the boxes**

```python
for a in ['x', 'y', 'z']:

    plt.figure(figsize = (10, 4), dpi = dpi, facecolor=(1, 1, 1))
    width = 0.25
    plt.bar(orig_assess.index - width/2, orig_assess[f'Mean - V{a}'], width, label = '640$^3$ h$^{-3}$Mpc$^3$')
    plt.bar(big_assess.index  + width/2, big_assess[f'Mean - V{a}'],  width, label = '1920$^3$ h$^{-3}$Mpc$^3$');
    plt.ylim(-1000,)
    plt.ylabel("Mean error in velocity (km/s)")
    plt.legend(loc = 'lower left')
    plt.xticks(ticks = orig_assess.index, labels = [f'C: {cell}, B: {box}' for cell, box in zip(orig_assess['Cell Size'], orig_assess['Act Cube Size'])], 
               rotation = 60, fontsize = 6, horizontalalignment = 'right');
    plt.xlabel('Cell and Box Sizes (h$^{-1}$ kpc)')
    plt.title("Original vs Big Simulation Box | Mean error for V$_{}$".format(a))

    net = big_assess[f'Mean - V{a}'] - orig_assess[f'Mean - V{a}']
    plt.text(0.75, 0.15, f'Net v vec +ve cases: {len(net[net>0])}', transform=plt.gca().transAxes)
    plt.text(0.75, 0.09, f'Net v vec -ve cases: {len(net[net<0])}', transform=plt.gca().transAxes)
    plt.text(0.75, 0.03, f'Average change: {round(net.mean(),1)} km/s', transform=plt.gca().transAxes)
    plt.tight_layout()
    plt.savefig(f'Plots/orig-vs-big-box/Signed_mean_of_error/V{a}.png')
```

**Conclusion**
1. To the credit of big simulation box, Pearson's increases for 67 out of 90 cases but only slightly: 0.003 (Vx), 0.006 (Vy), 0.007 (Vz) on average. 
2. But to its discredit, standard deviation in the velocity errors also increases in 69 out of 90 cases: 10 km/s (Vx), 2 km/s (Vy) & 3 km/s (Vz) on average.
3. Also to it discredit, the absolute value of mean of the velocity errors also increases in 62 out of 90 cases: 35 km/s (Vx) & 27 km/s (Vy) on average; for Vz, the mean decreases, i.e., it's -0.7 km/s. The mean velocity error tend to get more negative in 72 out of 90 cases: -50 km/s (Vx) & -38 km/s; for Vz, the mean gets positive, i.e., it's 0.7 km/s.

Hence, if Pearson's r is the most important indicator of the goodness of the simulation box, the big simulation box should be used. Else, if standard deviation & Mean of the error are important, the original simulation box should be preferred.
