```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_incorrect = pd.read_csv('./data/slope_and_r_480000_incorrect_f.csv', index_col = 0)
df_incorrect
```

```python
df_correct = pd.read_csv('./data/slope_and_r_480000_correct_f.csv', index_col = 0)
df_correct
```

```python
df_correct_intercept_fix_correct = pd.read_csv('./data/slope_and_r_480000_correct_f_intercept_fixed.csv', index_col = 0)
df_correct_intercept_fix_correct
```

```python
plt.figure(dpi = 120)
x_axis = np.arange(0,6)
plt.plot(x_axis, df_incorrect['r - Vlos'], label = 'f = $\Omega_\mathrm{m} ^{0.545}$')
plt.plot(x_axis, df_correct['r - Vlos'], label = 'f = $\Omega_\mathrm{m} \mathrm{(z)} ^{0.545}$', 
        ls = 'dashed')
plt.legend()
plt.xlabel(r'Cell size (h$^{-1}$kpc)')
plt.ylabel('Pearson\'s r')
plt.xticks(x_axis, [2000, 4000, 5000, 8000, 10000, 20000])
minus_one = {-1}
plt.title(r'Sphere radius: {} h$^{}$Mpc '.format(int(df_correct['Cube Size'][0]/2e3), minus_one));

plt.tight_layout()
plt.savefig('Plots/r_trend_for_cells.pdf')
```

```python
plt.figure(dpi = 120)
x_axis = np.arange(0,6)
plt.plot(x_axis, df_incorrect['Slope - Vlos'], label = 'f = $\Omega_\mathrm{m} ^{0.545}$')
plt.plot(x_axis, df_correct['Slope - Vlos'], label = 'f = $\Omega_\mathrm{m}  \mathrm{(z)} ^{0.545}$')
plt.plot(x_axis, df_correct_intercept_fix_correct['Slope - Vlos'], 
         label = 'f = $\Omega_\mathrm{m}  \mathrm{(z)} ^{0.545}$ | c = 0')
plt.legend()
plt.xlabel(r'Cell size (h$^{-1}$kpc)')
plt.ylabel('Slope of fit')
plt.xticks(x_axis, [2000, 4000, 5000, 8000, 10000, 20000])
minus_one = {-1}
plt.title(r'Sphere radius: {} h$^{}$Mpc '.format(int(df_correct['Cube Size'][0]/2e3), minus_one));

plt.tight_layout()
plt.savefig('Plots/slope_trend_for_cells.pdf')
```

```python
df_cell_fix_incorrect = pd.read_csv('./data/slope_and_r_cell_10000_incorrect_f.csv', index_col = 0)
df_cell_fix_incorrect
```

```python
df_cell_fix_correct = pd.read_csv('./data/slope_and_r_cell_10000_correct_f.csv', index_col = 0)
df_cell_fix_correct
```

```python
df_cell_intercept_fix_correct = pd.read_csv('./data/slope_and_r_cell_10000_correct_f_intercept_fixed.csv', index_col = 0)
df_cell_intercept_fix_correct
```

```python
plt.figure(dpi = 120)
x_axis = np.arange(0,6)
plt.plot(df_cell_fix_incorrect['Cube Size']/2e3, df_cell_fix_incorrect['r - Vlos'], label = 'f = $\Omega_\mathrm{m} ^{0.545}$')
plt.plot(df_cell_fix_correct['Cube Size']/2e3, df_cell_fix_correct['r - Vlos'], label = 'f = $\Omega_\mathrm{m} \mathrm{(z)} ^{0.545}$', 
        ls = 'dashed')
plt.legend()
plt.xlabel(r'Radius of sphere (h$^{-1}$Mpc)')
plt.ylabel('Pearson\'s r')
minus_one = {-1}
plt.title(r'Cell radius: {} h$^{}$Mpc '.format(int(df_cell_fix_incorrect['Cell Size'][0]/1e3), minus_one));

plt.tight_layout()
plt.savefig('Plots/r_trend_for_spheres.pdf')
```

```python
plt.figure(dpi = 120)
x_axis = np.arange(0,6)
plt.plot(df_cell_fix_incorrect['Cube Size']/2e3, df_cell_fix_incorrect['Slope - Vlos'], 
         label = 'f = $\Omega_\mathrm{m} ^{0.545}$')
plt.plot(df_cell_fix_correct['Cube Size']/2e3, df_cell_fix_correct['Slope - Vlos'], 
         label = 'f = $\Omega_\mathrm{m}  \mathrm{(z)} ^{0.545}$')
plt.plot(df_cell_intercept_fix_correct['Cube Size']/2e3, df_cell_intercept_fix_correct['Slope - Vlos'], 
         label = 'f = $\Omega_\mathrm{m} \mathrm{(z)} ^{0.545}$ | c = 0')
plt.legend()
plt.xlabel(r'Radius of sphere (h$^{-1}$Mpc)')
plt.ylabel('Slope of fit')
minus_one = {-1}
plt.title(r'Cell radius: {} h$^{}$Mpc '.format(int(df_cell_fix_incorrect['Cell Size'][0]/1e3), minus_one));

plt.tight_layout()
plt.savefig('Plots/slope_trend_for_spheres.pdf')
```
